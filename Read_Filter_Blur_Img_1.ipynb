```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("Dog.jpg")
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
import numpy as np
k1 = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
filtered_img = cv2.filter2D(img,-1,k1)
plt.imshow(filtered_img)
```

```python
blur_img = cv2.blur(img,(2,2))
plt.imshow(blur_img)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(new_img)
plt.title('Original Image')
plt.axis('off')

# Plotting the filtered image
plt.subplot(1, 3, 2)
plt.imshow(filtered_img)
plt.title('Filtered Image')
plt.axis('off')

# Plotting the blurred image
plt.subplot(1, 3, 3)
plt.imshow(blur_img)
plt.title('Blurred Image')
plt.axis('off')



plt.tight_layout()
plt.show()
```
